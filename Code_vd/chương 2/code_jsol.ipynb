```python
import json 
 
# Hàm kiểm tra chuỗi JSON có chứa đối tượng phức (complex object) hay không 
def is_complex(object): 
    # Kiểm tra xem trong đối tượng có tồn tại khóa '__complex__' hay không 
    if '__complex__' in object: 
        # Nếu có, tạo một số phức từ thông tin trong đối tượng 
        return complex(object['real'], object['img']) 
    # Nếu không có khóa '__complex__', đối tượng không phải là đối tượng phức,  
          # trả về đối tượng gốc 
    return object 
 
def main(): 
    # Sử dụng hàm json.loads() với object_hook để kiểm tra và chuyển đổi đối tượng       
          #phức trong chuỗi JSON 
    complex_object = json.loads('{"__complex__": true, "real": 4, "img": 5}',   
                               object_hook=is_complex) 
    simple_object = json.loads('{"real": 6, "img": 7}',  
                                object_hook=is_complex) 
 
    # In kết quả phân tích chuỗi JSON cho cả đối tượng phức và đối tượng thường 
    print("Đối tượng số phức......", complex_object) 
    print("Đối tượng bình thường......", simple_object) 
 
if __name__ == "__main__": 
    main() 
```

```python
import json 
# Định nghĩa một lớp đối tượng tùy chỉnh 
class CustomObject: 
    def __init__(self, data): 
        self.data = data 
    def to_dict(self): 
        return {"custom_data": self.data} 
# Tạo một lớp JSONEncoder tùy chỉnh để xử lý đối tượng CustomObject 
class CustomObjectEncoder(json.JSONEncoder): 
    def default(self, obj): 
        if isinstance(obj, CustomObject): 
            return obj.to_dict()  
        return super().default(obj) 

# Tạo một đối tượng CustomObject 
custom_obj = CustomObject({'key': 'value'}) 
# Sử dụng JSONEncoder tùy chỉnh để mã hóa đối tượng CustomObject 
json_data = json.dumps(custom_obj, cls=CustomObjectEncoder) 
print(json_data)
```

```python
# Import lớp JSONEncoder từ module json.encoder 
from json.encoder import JSONEncoder 
 
# Định nghĩa từ điển colour_dict chứa một danh sách các màu. 
colour_dict = { "colour": ["red", "yellow", "green" ]} 
 
# Sử dụng phương thức encode() của JSONEncoder để mã hóa từ điển colour_dict   
# thành chuỗi JSON. 
json_string=JSONEncoder().encode(colour_dict) 
 
#in chuỗi json_string được mã hóa ra màn hình. 
print(json_string) 
 
#in ra kiểu của chuỗi json_string 
print(type(json_string))
```

```python
import json 
# Import lớp JSONDecoder từ module json.decoder 
from json.decoder import JSONDecoder 
 
# Chuỗi JSON chứa dữ liệu màu sắc. 
colour_string = '{ "colour": ["red", "yellow"]}' 
 
# Sử dụng JSONDecoder để giải mã chuỗi JSON thành đối tượng Python. 
# JSONDecoder là một lớp con của json.JSONDecoder, cho phép tùy chỉnh quá trình 
#giải mã. 
# Phương thức decode() sẽ giải mã chuỗi JSON và trả về đối tượng Python tương ứng. 
decoded_object = JSONDecoder().decode(colour_string) 
 
# In đối tượng Python đã giải mã và kiểu của đối tượng đó ra màn hình. 
print(decoded_object) 
print(type(decoded_object)) 
```
